```python
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# -------------------------
# 1. Define project phases
# -------------------------
phases = [
    {
        "name": "Phase 1 – Foundation Development",
        "start": dt.date(2023, 12, 1),
        "end":   dt.date(2024, 4, 30),
    },
    {
        "name": "Phase 2 – Algorithm Expansion",
        "start": dt.date(2024, 5, 1),
        "end":   dt.date(2025, 4, 30),
    },
    {
        "name": "Phase 3 – Cross-Dataset Integration",
        "start": dt.date(2025, 5, 1),
        "end":   dt.date(2025, 12, 31),
    },
    {
        "name": "Phase 4 – Summer 2026 URC Research",
        "start": dt.date(2026, 5, 1),
        "end":   dt.date(2026, 8, 31),
    },
]

# -------------------------
# 2. Convert to matplotlib dates
# -------------------------
start_dates = [mdates.date2num(p["start"]) for p in phases]
durations = [(p["end"] - p["start"]).days for p in phases]
labels = [p["name"] for p in phases]

# y-positions for horizontal bars (0, 1, 2, 3)
y_pos = list(range(len(phases)))

# -------------------------
# 3. Create the figure
# -------------------------
plt.figure(figsize=(10, 3.5))

# Horizontal bars = timeline
plt.barh(
    y_pos,
    durations,
    left=start_dates,
    height=0.5,
    align="center"
)

# -------------------------
# 4. Format axes
# -------------------------
ax = plt.gca()

# Y-axis: phase labels
ax.set_yticks(y_pos)
ax.set_yticklabels(labels, fontsize=9)

# X-axis: dates
ax.xaxis_date()
ax.xaxis.set_major_locator(mdates.YearLocator())          # main ticks: years
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

# Optional: minor ticks for half-years or quarters
ax.xaxis.set_minor_locator(mdates.MonthLocator(interval=3))

plt.xticks(rotation=30, ha='right', fontsize=9)

# Invert y-axis so Phase 1 is at the top
ax.invert_yaxis()

# Labels and title
plt.title("URC Project Timeline – Bias-Aware Recommender Systems", fontsize=11)
plt.xlabel("Time")

plt.tight_layout()

# -------------------------
# 5. Save the figure
# -------------------------
plt.savefig("urc_timeline.png", dpi=600, bbox_inches="tight")
plt.savefig("urc_timeline.pdf", bbox_inches="tight")

plt.show()
```
