```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

# =========================
# Paths
# =========================

original_file = "/home/moshtasa/Research/phd-svd-recsys/Mobile/0316_Single/data/data/mobilerec_final.csv"
mapped_file = "/home/moshtasa/Research/phd-svd-recsys/Mobile/0316_Single/data/data/mobilerec_final_mapped.csv"

OUT_DIR = Path("/home/moshtasa/Research/phd-svd-recsys/Mobile/0316_Single/domain_analysis/figures")
OUT_DIR.mkdir(parents=True, exist_ok=True)

# =========================
# Load original dataset
# =========================

df = pd.read_csv(original_file, low_memory=False)

print("Dataset loaded")
print(df.shape)

# =========================
# Category Mapping
# =========================

category_mapping = {

# Games
"Action":"Games",
"Puzzle":"Games",
"Simulation":"Games",
"Strategy":"Games",
"Casual":"Games",
"Adventure":"Games",
"Sports":"Games",
"Arcade":"Games",
"Casino":"Games",
"Card":"Games",
"Board":"Games",
"Word":"Games",
"Racing":"Games",
"Role Playing":"Games",
"Trivia":"Games",

# Education
"Education":"Education & Learning",
"Educational":"Education & Learning",
"Books & Reference":"Education & Learning",
"Libraries & Demo":"Education & Learning",

# Health
"Health & Fitness":"Health & Lifestyle",
"Lifestyle":"Health & Lifestyle",
"Beauty":"Health & Lifestyle",
"Parenting":"Health & Lifestyle",

# Media
"Entertainment":"Media & Entertainment",
"Music":"Media & Entertainment",
"Music & Audio":"Media & Entertainment",
"Video Players & Editors":"Media & Entertainment",
"Comics":"Media & Entertainment",

# Social
"Social":"Social & Communication",
"Communication":"Social & Communication",
"Dating":"Social & Communication",

# Shopping
"Shopping":"Shopping & Food",
"Food & Drink":"Shopping & Food",

# Travel
"Travel & Local":"Travel & Navigation",
"Maps & Navigation":"Travel & Navigation",
"Auto & Vehicles":"Travel & Navigation",

# Productivity
"Productivity":"Work & Productivity",
"Business":"Work & Productivity",
"Finance":"Work & Productivity",
"Tools":"Work & Productivity",

# Utilities
"Weather":"Utilities & Info",
"News & Magazines":"Utilities & Info",
"Photography":"Utilities & Info",
"Art & Design":"Utilities & Info",
"Personalization":"Utilities & Info",
"House & Home":"Utilities & Info",
"Events":"Utilities & Info"
}

# =========================
# Create mapped dataset
# =========================

df_final_mapped = df.copy()
df_final_mapped["category_mapped"] = df_final_mapped["app_category"].map(category_mapping)

# =========================
# Check unmapped categories
# =========================

missing = df_final_mapped["category_mapped"].isna().sum()
print("\nUnmapped categories:", missing)

if missing > 0:
    print(df_final_mapped[df_final_mapped["category_mapped"].isna()]["app_category"].unique())

# =========================
# Save mapped dataset
# =========================

df_final_mapped.to_csv(mapped_file, index=False)

print("\nSaved mapped dataset:")
print(mapped_file)

# =========================
# CATEGORY COUNTS (Original)
# =========================

original_counts = (
    df[["app_package","app_category"]]
    .drop_duplicates()
    ["app_category"]
    .value_counts()
    .sort_values(ascending=False)
)

original_counts.to_csv(OUT_DIR / "original_category_counts.csv")

# =========================
# CATEGORY COUNTS (Mapped)
# =========================

mapped_counts = (
    df_final_mapped[["app_package","category_mapped"]]
    .drop_duplicates()
    ["category_mapped"]
    .value_counts()
    .sort_values(ascending=False)
)

mapped_counts.to_csv(OUT_DIR / "mapped_category_counts.csv")

# =========================
# Plot helper (UPDATED STYLE)
# =========================

def create_bar_chart(data, title, filename):

    x = np.arange(len(data)) * 1.4
    width = 0.7

    fig, ax = plt.subplots(figsize=(14,7))

    bars = ax.bar(
        x,
        data.values,
        width=width,
        color="#5A6F80",   # single mild blue-gray color
        edgecolor="black",
        linewidth=1.2
    )

    # =========================
    # STYLE (Figure 1 style)
    # =========================

    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    ax.spines["bottom"].set_linewidth(1.2)
    ax.spines["left"].set_linewidth(1.0)

    ax.spines["bottom"].set_position(("outward", 5))

    ax.yaxis.grid(True, linestyle="--", linewidth=0.5, alpha=0.5)
    ax.xaxis.grid(False)

    # =========================

    ax.set_title(title, fontsize=14)
    ax.set_ylabel("Number of Apps")

    ax.set_xticks(x)
    ax.set_xticklabels(data.index, rotation=90)

    # NOTE: removed value labels on bars

    plt.tight_layout()
    plt.savefig(OUT_DIR / filename, dpi=300)
    plt.close()

# =========================
# BAR CHART (Before mapping)
# =========================

create_bar_chart(
    original_counts,
    "Original Mobile App Categories (48 Categories)",
    "fig_original_categories_bar.png"
)

# =========================
# BAR CHART (After mapping)
# =========================

create_bar_chart(
    mapped_counts,
    "Mapped Mobile App Domains (9 Categories)",
    "fig_mapped_categories_bar.png"
)

# =========================
# Final statistics
# =========================

print("\nOriginal Categories:", original_counts.shape[0])
print("Mapped Categories:", mapped_counts.shape[0])

print("\nUnique users:", df["uid"].nunique())
print("Unique apps:", df["app_package"].nunique())

print("\nOutputs saved to:")
print(OUT_DIR)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

# =========================
# Paths
# =========================

original_file = "/home/moshtasa/Research/phd-svd-recsys/Mobile/0316_Single/data/data/mobilerec_final.csv"
mapped_file = "/home/moshtasa/Research/phd-svd-recsys/Mobile/0316_Single/data/data/mobilerec_final_mapped.csv"

OUT_DIR = Path("/home/moshtasa/Research/phd-svd-recsys/Mobile/0316_Single/domain_analysis/figures")
OUT_DIR.mkdir(parents=True, exist_ok=True)

# =========================
# Load dataset
# =========================

df = pd.read_csv(original_file, low_memory=False)

# =========================
# Category Mapping
# =========================

category_mapping = {
"Action":"Games","Puzzle":"Games","Simulation":"Games","Strategy":"Games","Casual":"Games",
"Adventure":"Games","Sports":"Games","Arcade":"Games","Casino":"Games","Card":"Games",
"Board":"Games","Word":"Games","Racing":"Games","Role Playing":"Games","Trivia":"Games",

"Education":"Education & Learning","Educational":"Education & Learning",
"Books & Reference":"Education & Learning","Libraries & Demo":"Education & Learning",

"Health & Fitness":"Health & Lifestyle","Lifestyle":"Health & Lifestyle",
"Beauty":"Health & Lifestyle","Parenting":"Health & Lifestyle",

"Entertainment":"Media & Entertainment","Music":"Media & Entertainment",
"Music & Audio":"Media & Entertainment","Video Players & Editors":"Media & Entertainment",
"Comics":"Media & Entertainment",

"Social":"Social & Communication","Communication":"Social & Communication",
"Dating":"Social & Communication",

"Shopping":"Shopping & Food","Food & Drink":"Shopping & Food",

"Travel & Local":"Travel & Navigation","Maps & Navigation":"Travel & Navigation",
"Auto & Vehicles":"Travel & Navigation",

"Productivity":"Work & Productivity","Business":"Work & Productivity",
"Finance":"Work & Productivity","Tools":"Work & Productivity",

"Weather":"Utilities & Info","News & Magazines":"Utilities & Info",
"Photography":"Utilities & Info","Art & Design":"Utilities & Info",
"Personalization":"Utilities & Info","House & Home":"Utilities & Info",
"Events":"Utilities & Info"
}

df["category_mapped"] = df["app_category"].map(category_mapping)

# =========================
# Counts
# =========================

original_counts = (
    df[["app_package","app_category"]]
    .drop_duplicates()["app_category"]
    .value_counts()
    .sort_values(ascending=False)
)

mapped_counts = (
    df[["app_package","category_mapped"]]
    .drop_duplicates()["category_mapped"]
    .value_counts()
    .sort_values(ascending=False)
)

# =========================
# Create side-by-side plots
# =========================

fig, axes = plt.subplots(1, 2, figsize=(6.5, 2.6))

# =========================
# LEFT: ORIGINAL (LINE PLOT)
# =========================

spacing = 0.9
x1 = np.arange(len(original_counts)) * spacing

axes[0].plot(
    x1,
    original_counts.values,
    color="#6A4C93",
    linewidth=1.2,
    marker="o",
    markersize=2
)

axes[0].spines["top"].set_visible(False)
axes[0].spines["right"].set_visible(False)
axes[0].spines["bottom"].set_linewidth(1.0)
axes[0].spines["left"].set_linewidth(0.8)

axes[0].yaxis.grid(True, linestyle="--", linewidth=0.4, alpha=0.5)

axes[0].set_title("Original Categories", fontsize=8)
axes[0].set_xlabel("Category", fontsize=7)
axes[0].set_ylabel("Number of Unique Applications", fontsize=7)

axes[0].set_xticks(x1)
axes[0].set_xticklabels(original_counts.index, rotation=75, fontsize=5)

axes[0].tick_params(axis='y', labelsize=6)

# =========================
# RIGHT: MAPPED (BAR PLOT)
# =========================

spacing = 1.1
x2 = np.arange(len(mapped_counts)) * spacing

axes[1].bar(
    x2,
    mapped_counts.values,
    width=0.35,
    color="#E6C200",
    edgecolor="black",
    linewidth=1.0
)

axes[1].spines["top"].set_visible(False)
axes[1].spines["right"].set_visible(False)
axes[1].spines["bottom"].set_linewidth(1.0)
axes[1].spines["left"].set_linewidth(0.8)

axes[1].yaxis.grid(True, linestyle="--", linewidth=0.4, alpha=0.5)

axes[1].set_title("Mapped Domains", fontsize=8)
axes[1].set_xlabel("Category", fontsize=7)
axes[1].set_ylabel("Number of Unique Applications", fontsize=7)

axes[1].set_xticks(x2)
axes[1].set_xticklabels(mapped_counts.index, rotation=45, fontsize=6)

axes[1].tick_params(axis='y', labelsize=6)

# =========================
# Save
# =========================

plt.tight_layout(pad=0.6, w_pad=1.0)
plt.savefig(OUT_DIR / "fig_combined_categories.png", dpi=300)
plt.close()

print("Saved combined figure to:")
print(OUT_DIR)
```
