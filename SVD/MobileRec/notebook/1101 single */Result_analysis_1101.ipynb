```python
#!/usr/bin/env python3
# make_all_roots_explanations_and_figs_apps_1101_10roots.py
#
# Apps dataset (MobileRec) — SINGLE-ROOT mode with robust normalization for 10 known categories.
# Reads ORIGINAL Top-K from SVD_Original/5 and injections from SVD_Single_Root/5.
# Outputs per-root: explanation.txt, per_app_ranking.csv, and a POS=5 bar figure.
#
# Python 3.8+

from pathlib import Path
from typing import Optional, Dict, Tuple, List
import re
import unicodedata
import pandas as pd
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt

# ========= PATHS =============================================================
SEARCH_ROOT = Path("/home/moshtasa/Research/phd-svd-recsys/SVD/MobileRec/result/rec/top_re/1101/SVD_Single_Root/5")
ORIG_DIR    = Path("/home/moshtasa/Research/phd-svd-recsys/SVD/MobileRec/result/rec/top_re/1101/SVD_Original/5")
OUT_BASE    = SEARCH_ROOT / "_analysis"

K_LIST = [35]
# Just for legend ordering if present; actual available Ns are discovered on disk per root.
N_LIST_HINT = [27, 105, 2643, 2, 4, 6, 25, 50, 100, 200, 300, 350, 500, 1000]

# ========= CANONICALIZATION (10 roots) ======================================
# Canonical keys (lowercase, space-separated)
TEN_CANON_KEYS = {
    "business and finance",
    "games",
    "health and lifestyle",
    "information and reading",
    "learning and education",
    "media and creativity",
    "medical",
    "productivity and tools",
    "social and communications",
    "travel and navigation",
}

# Display names as you likely want them shown
CANON_DISPLAY = {
    "business and finance":       "Business & Finance",
    "games":                      "Games",
    "health and lifestyle":       "Health & Lifestyle",
    "information and reading":    "Information & Reading",
    "learning and education":     "Learning & Education",
    "media and creativity":       "Media & Creativity",
    "medical":                    "Medical",
    "productivity and tools":     "Productivity & Tools",
    "social and communications":  "Social & Communications",
    "travel and navigation":      "Travel & Navigation",
}

# Alias table collapses common raw variants (filenames & CSV item_root) into canonical keys.
ALIAS_TO_CANON = {
    # Business & Finance
    "business & finance": "business and finance",
    "business and finance": "business and finance",
    "business finance": "business and finance",
    "business   finance": "business and finance",
    "business / finance": "business and finance",
    "business___finance": "business and finance",
    "business_finance": "business and finance",

    # Games
    "games": "games",

    # Health & Lifestyle
    "health & lifestyle": "health and lifestyle",
    "health and lifestyle": "health and lifestyle",
    "health lifestyle": "health and lifestyle",
    "health___lifestyle": "health and lifestyle",
    "health_lifestyle": "health and lifestyle",

    # Information & Reading
    "information & reading": "information and reading",
    "information and reading": "information and reading",
    "information reading": "information and reading",
    "info & reading": "information and reading",
    "info and reading": "information and reading",
    "info reading": "information and reading",
    "info___reading": "information and reading",
    "information___reading": "information and reading",
    "info_reading": "information and reading",

    # Learning & Education
    "learning & education": "learning and education",
    "learning and education": "learning and education",
    "learning education": "learning and education",
    "learning___education": "learning and education",
    "learning_education": "learning and education",

    # Media & Creativity
    "media & creativity": "media and creativity",
    "media and creativity": "media and creativity",
    "media creativity": "media and creativity",
    "media___creativity": "media and creativity",
    "media_creativity": "media and creativity",

    # Medical
    "medical": "medical",

    # Productivity & Tools
    "productivity & tools": "productivity and tools",
    "productivity and tools": "productivity and tools",
    "productivity tools": "productivity and tools",
    "productivity___tools": "productivity and tools",
    "productivity_tools": "productivity and tools",

    # Social & Communications
    "social & communications": "social and communications",
    "social and communications": "social and communications",
    "social & comms": "social and communications",
    "social and comms": "social and communications",
    "social communications": "social and communications",
    "social communication": "social and communications",
    "social___comms": "social and communications",
    "social_comms": "social and communications",

    # Travel & Navigation
    "travel & navigation": "travel and navigation",
    "travel and navigation": "travel and navigation",
    "travel navigation": "travel and navigation",
    "travel___navigation": "travel and navigation",
    "travel_navigation": "travel and navigation",
}

def _basic_clean(s: str) -> str:
    s = unicodedata.normalize("NFKC", str(s))
    s = s.replace("’", "'").replace("“", '"').replace("”", '"')
    s = s.replace("_", " ")
    s = re.sub(r"[\/\-]+", " ", s)          # slashes/hyphens -> space
    s = re.sub(r"&", " and ", s, flags=re.I) # ampersand -> and
    s = re.sub(r"[^\w\s']", " ", s)          # drop other punctuation (keep apostrophes)
    s = re.sub(r"\s+", " ", s).strip()
    return s

def canon_key(s: str) -> str:
    """Canonicalize any category token into one of TEN_CANON_KEYS (if possible)."""
    if s is None or str(s).strip() == "":
        return ""
    base = _basic_clean(s).lower()
    # quick alias hit
    if base in ALIAS_TO_CANON:
        return ALIAS_TO_CANON[base]
    # fallbacks for close variants:
    # map "comms"→"communications", "info"→"information"
    base2 = re.sub(r"\bcomms\b", "communications", base)
    base2 = re.sub(r"\binfo\b", "information", base2)
    # normalize multiple spaces
    base2 = re.sub(r"\s+", " ", base2).strip()
    # if "and" missing between two words that are known pairs, insert "and" (heuristic)
    pairs = [
        ("business", "finance"),
        ("health", "lifestyle"),
        ("information", "reading"),
        ("learning", "education"),
        ("media", "creativity"),
        ("productivity", "tools"),
        ("social", "communications"),
        ("travel", "navigation"),
    ]
    for a, b in pairs:
        if base2 == f"{a} {b}" or base2 == f"{a}  {b}":
            base2 = f"{a} and {b}"
    # Final mapping if in our 10 keys
    if base2 in TEN_CANON_KEYS:
        return base2
    # Try alias again (e.g., with inserted "and")
    if base2 in ALIAS_TO_CANON:
        return ALIAS_TO_CANON[base2]
    # As last resort, if it starts with one of our 10 keys (rare naming with suffixes), trim
    for key in TEN_CANON_KEYS:
        if base2.startswith(key):
            return key
    return base2  # left as-is; may be filtered out later

def display_name(canon: str) -> str:
    return CANON_DISPLAY.get(canon, canon.title())

# ========= File patterns / parsing ===========================================
# f_<Root>_{N}u_pos5_negX_all_{K}recommendation.csv
INJECT_RE = re.compile(r"^f_(.+?)_(\d+)u_pos\d+_neg[^_]+_all_(\d+)recommendation\.csv$", re.IGNORECASE)

def extract_injection_info(p: Path):
    m = INJECT_RE.match(p.name)
    if not m:
        return None
    raw_root = m.group(1)  # may contain underscores or triple underscores
    n = int(m.group(2))
    k = int(m.group(3))
    return canon_key(raw_root), n, k

def discover_roots_and_ns() -> Tuple[List[str], Dict[str, set]]:
    """Scan SEARCH_ROOT to find all canonical roots present and Ns per root."""
    roots_canon = set()
    ns_by_root: Dict[str, set] = {}
    for p in SEARCH_ROOT.rglob("f_*_pos5_*recommendation.csv"):
        info = extract_injection_info(p)
        if not info:
            continue
        root_can, n, _k = info
        if not root_can:
            continue
        root_can = canon_key(root_can)
        # restrict to the 10 roots we expect
        if root_can not in TEN_CANON_KEYS:
            continue
        roots_canon.add(root_can)
        ns_by_root.setdefault(root_can, set()).add(n)
    return sorted(roots_canon), ns_by_root

# ========= IO helpers ========================================================
def load_rec_csv(fp: Path) -> pd.DataFrame:
    df = pd.read_csv(fp, low_memory=False)
    for c in ["user_id", "app_id", "rank", "est_score", "item_root", "item_subroot"]:
        if c not in df.columns:
            df[c] = pd.NA
    df["_root_canon"] = df["item_root"].astype(str).map(canon_key)
    return df

def metrics_for_file(df: pd.DataFrame, target_root_canon: str):
    unique_apps_in_file = df["app_id"].nunique() if "app_id" in df.columns else 0
    is_tg = (df["_root_canon"] == target_root_canon)
    if "user_id" in df.columns:
        per_user = df.assign(is_tg=is_tg).groupby("user_id")["is_tg"].sum()
        avg_per_user = float(per_user.mean()) if not per_user.empty else 0.0
        users_with_target = int((per_user > 0).sum())
    else:
        avg_per_user = 0.0
        users_with_target = 0
    unique_tg_apps = df.loc[is_tg, "app_id"].nunique() if "app_id" in df.columns else 0
    freq = int(is_tg.sum())
    return unique_apps_in_file, avg_per_user, unique_tg_apps, freq, users_with_target, is_tg

def per_app_ranking(df_target_rows: pd.DataFrame) -> pd.DataFrame:
    if df_target_rows.empty:
        return pd.DataFrame(columns=[
            "app_id","rank","freq","users_n","avg_rank","avg_est_score","rank1_count","item_root","item_subroot"
        ])
    g = df_target_rows.groupby("app_id", as_index=False)
    out = g.agg(
        freq=("app_id", "size"),
        users_n=("user_id", "nunique"),
        avg_rank=("rank", "mean"),
        avg_est_score=("est_score", "mean"),
        rank1_count=("rank", lambda s: int((s == 1).sum())),
    )
    for c in ["item_root", "item_subroot"]:
        smpl = df_target_rows.groupby("app_id")[c].apply(
            lambda s: s.dropna().iloc[0] if s.notna().any() else pd.NA
        ).reset_index(name=c)
        out = out.merge(smpl, on="app_id", how="left")
    out = out.sort_values(["freq","rank1_count","avg_rank"], ascending=[False, False, True]).reset_index(drop=True)
    out.insert(1, "rank", out.index + 1)
    return out

def root_key_and_disp(root_canon: str) -> Tuple[str, str]:
    disp = display_name(root_canon)
    key  = re.sub(r"[^\w\-]+", "_", disp)
    return key, disp

# ========= Plotting ==========================================================
def plot_root_pos_three_bins(R_disp: str,
                             data_by_k: Dict[int, Dict[str, float]],
                             out_png: Path,
                             present_groups: List[str]):
    ks = sorted(data_by_k.keys())
    if not ks:
        return
    width = 0.8 / max(1, len(present_groups))
    x = list(range(len(ks)))
    plt.figure(figsize=(8.4, 4.4), dpi=160)
    for j, g in enumerate(present_groups):
        offs = [i + (j - (len(present_groups)-1)/2)*width for i in x]
        vals = [float(data_by_k.get(k, {}).get(g, 0.0)) for k in ks]
        plt.bar(offs, vals, width=width, label=("n="+g if g!="Original" else "Original"))
    plt.xticks(x, [f"Top-{k}" for k in ks])
    plt.ylabel("Avg # of apps in Top-K per user (target root)")
    plt.title(f"{R_disp} — POS=5")
    plt.legend(ncol=min(4, len(present_groups)), fontsize=9)
    plt.tight_layout()
    out_png.parent.mkdir(parents=True, exist_ok=True)
    plt.savefig(out_png)
    plt.close()

# ========= MAIN per-root =====================================================
def process_one_root(target_root_canon: str, ns_for_this_root: List[int]) -> Dict:
    ROOT_KEY, ROOT_DISP = root_key_and_disp(target_root_canon)
    OUT_DIR = OUT_BASE / f"{ROOT_KEY}_explanation"
    OUT_DIR.mkdir(parents=True, exist_ok=True)

    original_uniques: Dict[int, int] = {}
    injection_uniques: Dict[Tuple[int,int], int] = {}
    per_app_rows: List[Dict] = []
    data_by_k: Dict[int, Dict[str, float]] = {k: {} for k in K_LIST}

    # ORIGINAL
    for K in K_LIST:
        fp = ORIG_DIR / f"ORIGINAL_{K}recommendation.csv"
        if not fp.exists():
            print(f"[warn][{ROOT_DISP}] Missing ORIGINAL for K={K}: {fp}")
            continue
        df = load_rec_csv(fp)
        uniq_all, avg_user, uniq_tg, freq, users_with_tg, is_tg = metrics_for_file(df, target_root_canon)
        print(f"[{ROOT_DISP}] [{K}], {fp.name} -> {uniq_all} apps -> {ROOT_DISP}: avg/user={avg_user:.2f}, unique={uniq_tg}")
        original_uniques[K] = uniq_tg
        data_by_k[K]["Original"] = avg_user

        ranked = per_app_ranking(df.loc[is_tg].copy())
        for _, r in ranked.iterrows():
            per_app_rows.append({
                "root": ROOT_DISP,
                "dataset": f"original{K}",
                "app_id": int(r["app_id"]),
                "rank": int(r["rank"]),
                "freq": int(r["freq"]),
                "users_n": int(r["users_n"]),
                "avg_rank": float(r["avg_rank"]) if pd.notna(r["avg_rank"]) else None,
                "avg_est_score": float(r["avg_est_score"]) if pd.notna(r["avg_est_score"]) else None,
                "rank1_count": int(r["rank1_count"]),
                "item_root": r.get("item_root", pd.NA),
                "item_subroot": r.get("item_subroot", pd.NA),
            })

    # INJECTIONS
    ns_sorted = sorted(set(ns_for_this_root) | set(N_LIST_HINT), key=lambda x: (x not in ns_for_this_root, x))
    any_group_present = set()
    for K in K_LIST:
        for N in ns_sorted:
            patt = f"f_*_{N}u_pos5_*_{K}recommendation.csv"
            found_fp: Optional[Path] = None
            for p in SEARCH_ROOT.rglob(patt):
                info = extract_injection_info(p)
                if not info:
                    continue
                r_can, n_found, k_found = info
                if r_can == target_root_canon and n_found == N and k_found == K:
                    found_fp = p
                    break
            if not found_fp:
                continue

            df = load_rec_csv(found_fp)
            uniq_all, avg_user, uniq_tg, freq, users_with_tg, is_tg = metrics_for_file(df, target_root_canon)
            print(f"[{ROOT_DISP}] [{K}] {found_fp.name} -> {uniq_all} apps -> {ROOT_DISP}: avg/user={avg_user:.2f}, unique={uniq_tg}")
            injection_uniques[(N, K)] = uniq_tg
            data_by_k[K][str(N)] = avg_user
            any_group_present.add(str(N))

            ranked = per_app_ranking(df.loc[is_tg].copy())
            for _, r in ranked.iterrows():
                per_app_rows.append({
                    "root": ROOT_DISP,
                    "dataset": f"{N}u_{K}",
                    "app_id": int(r["app_id"]),
                    "rank": int(r["rank"]),
                    "freq": int(r["freq"]),
                    "users_n": int(r["users_n"]),
                    "avg_rank": float(r["avg_rank"]) if pd.notna(r["avg_rank"]) else None,
                    "avg_est_score": float(r["avg_est_score"]) if pd.notna(r["avg_est_score"]) else None,
                    "rank1_count": int(r["rank1_count"]),
                    "item_root": r.get("item_root", pd.NA),
                    "item_subroot": r.get("item_subroot", pd.NA),
                })

    # Write per-root text
    text_path = OUT_DIR / "explanation.txt"
    with open(text_path, "w", encoding="utf-8") as f:
        f.write(f"{ROOT_DISP.lower()}:\n")
        for K in K_LIST:
            if K in original_uniques:
                f.write(f"original {K}: number_of_unique_apps: {original_uniques[K]}\n")
        for K in K_LIST:
            for N in sorted({n for (n,k) in injection_uniques.keys() if k == K}):
                val = injection_uniques.get((N, K))
                if val is not None:
                    f.write(f"{N}u, {K}, number_of_unique_apps: {val}\n")

    # Write per-root ranking table
    table_df = pd.DataFrame(per_app_rows)
    table_path = OUT_DIR / "per_app_ranking.csv"
    table_df.to_csv(table_path, index=False)

    # Make and save per-root figure
    present_groups = ["Original"] + sorted(any_group_present, key=lambda x: (x not in [str(n) for n in N_LIST_HINT], int(x)))
    fig_path = OUT_DIR / f"{re.sub(r'[^A-Za-z0-9]+', '_', ROOT_DISP).strip('_').lower()}__pos5.png"
    plot_root_pos_three_bins(ROOT_DISP, data_by_k, fig_path, present_groups)

    print(f"[OK][{ROOT_DISP}] Saved text:   {text_path}")
    print(f"[OK][{ROOT_DISP}] Saved table:  {table_path}")
    print(f"[OK][{ROOT_DISP}] Saved figure: {fig_path}")

    return {
        "disp": ROOT_DISP,
        "original_uniques": original_uniques,
        "injection_uniques": injection_uniques,
        "out_dir": OUT_DIR,
        "per_app_rows": per_app_rows,
    }

# ========= MAIN (all roots) ==================================================
def main():
    OUT_ALL = OUT_BASE / "_all_roots"
    OUT_ALL.mkdir(parents=True, exist_ok=True)

    roots_found, ns_by_root = discover_roots_and_ns()
    if not roots_found:
        print("[warn] No roots discovered under injection dir. Nothing to do.")
        return

    # Keep only our 10 categories (defensive)
    roots_found = [r for r in roots_found if r in TEN_CANON_KEYS]

    all_rows: List[Dict] = []
    master_lines: List[str] = []

    for r_can in roots_found:
        res = process_one_root(r_can, sorted(ns_by_root.get(r_can, [])))
        master_lines.append(f"{res['disp'].lower()}:")
        for K in K_LIST:
            if K in res["original_uniques"]:
                master_lines.append(f"original {K}: number_of_unique_apps: {res['original_uniques'][K]}")
        for K in K_LIST:
            for (N, KK), val in sorted(res["injection_uniques"].items()):
                if KK == K:
                    master_lines.append(f"{N}u, {K}, number_of_unique_apps: {val}")
        master_lines.append("")  # blank line
        all_rows.extend(res["per_app_rows"])

    # Save global rollups
    master_txt = OUT_ALL / "summary_master.txt"
    with open(master_txt, "w", encoding="utf-8") as f:
        f.write("\n".join(master_lines))
    print(f"[OK] Wrote master summary: {master_txt}")

    if all_rows:
        df_all = pd.DataFrame(all_rows)
        df_all.to_csv(OUT_ALL / "per_app_ranking_all.csv", index=False)
        print(f"[OK] Wrote global per-app ranking CSV: {OUT_ALL / 'per_app_ranking_all.csv'}")
    else:
        print("[warn] No per-app rows produced. Check inputs / paths.")

if __name__ == "__main__":
    main()
```
