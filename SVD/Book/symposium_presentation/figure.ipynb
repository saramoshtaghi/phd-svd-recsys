```python
import matplotlib.pyplot as plt

# Data for K=15 (Adult)
synthetic_users = ["original", "25", "50", "100", "200"]
counts = [0.601939, 0.695717, 0.663541, 0.704403, 0.646357]

# Assign distinct colors for each bar
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd"]

# Create figure
plt.figure(figsize=(8, 5))

# Plot bars with assigned colors
bars = plt.bar(synthetic_users, counts, color=colors)

# Add value labels on top of bars
for bar, c in zip(bars, counts):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
             f"{c:.3f}", ha="center", va="bottom", fontsize=10, weight="bold")

# Labels and title
plt.ylabel("Number of books recommended in Adults")
plt.xlabel("Number of synthetic users")
plt.title("Adult â€” K=15")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from your dataset summary
categories = [
    'Total Rows',
    'Unique Users',
    'Unique Items',
    'Unique Decades',
    'Avg Ratings/User',
    'Min Ratings/User',
    'Max Ratings/User'
]
values = [100000, 943, 1682, 8, 106.04, 20, 737]
colors = ['#2E86AB', '#A23B72', '#F18F01', '#C73E1D', '#3C91E6', '#FF6F61', '#6A4C93']

# Create figure
fig, ax = plt.subplots(figsize=(7, 6))

# Vertical bar chart (log scale helps spread values across rows vs. decades/users)
bars = ax.bar(categories, values, color=colors, alpha=0.85, edgecolor='black', linewidth=1)

# Log scale for better visualization (since Total Rows >> others)
ax.set_yscale('log')
ax.set_ylim(1, 2000000)  # extend upper limit beyond 100k for clarity

# Add value labels
for i, (bar, value) in enumerate(zip(bars, values)):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height * 1.25,
            f'{value:,.2f}' if isinstance(value, float) else f'{value:,}',
            ha='center', va='bottom',
            fontsize=12, fontweight='bold')

# Styling
ax.set_ylabel('Count (Log Scale)', fontsize=14, fontweight='bold')
ax.set_title('Dataset Summary: User-Item Ratings Structure',
             fontsize=16, fontweight='bold', pad=20)

# Grid and formatting
ax.grid(axis='y', alpha=0.3, linestyle='--')
ax.set_axisbelow(True)
plt.xticks(fontsize=11, fontweight='bold', rotation=35, ha='right')
plt.yticks(fontsize=11, fontweight='bold')

# Adjust layout for labels
plt.subplots_adjust(top=0.85, bottom=0.2)

# Save high-resolution figure
plt.savefig("dataset_summary_scale.png", dpi=300, bbox_inches="tight")

# Show
plt.show()
```
